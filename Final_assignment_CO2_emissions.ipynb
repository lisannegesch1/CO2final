```python
import pandas as pd
import numpy as np
```

```python
# Data inlezen
# Annual CO2 Emissions
df_co2 = pd.read_csv("/content/annual-co2-emissions-per-country.csv", encoding = 'cp1252')

# population growth
df_pgr = pd.read_csv("/content/UN-population-projection-medium-variant.csv", encoding = 'cp1252')

df_co2
#df_pgr
```

```python
#df_co2.columns
df_pgr.columns
```

```python
# Data opschonen
# CO2
df_co2 = df_co2.drop(columns=['Code'])
df_co2 = df_co2.rename(columns={'Annual CO2 emissions': 'CO2'})
#df_pgr
df_pgr = df_pgr.drop(columns=['Code'])
df_pgr = df_pgr.rename(columns={'Population (future projections)': 'Future', 'Population (historical estimates)': 'Histroic'})

#data samenvoegen
df2 = pd.merge(df_co2, df_pgr, how="inner", on = ["Entity", "Year"])
```

```python
#opschonen voor correlatie
df3 = df2

df3 = df3.drop(columns=['Future'])
df3 = df3.set_index('Year')
df3
```

```python
correlation = df3.dropna().corr()
correlation
# uit het verleden blijkt dat er een hele sterke correlatie is met de groei van de populatie en de groei van de CO2 uitstoot
```

```python
# de voorspelling van het grootste land 
df_pgr2 = df_pgr.sort_values('Future', ascending=False)

values = ['World', 'Asia', 'Africa']
df_pgr3 = df_pgr2[df_pgr2.Entity.isin(values) == False]
#df_pgr3 = df_pgr2[df_pgr2.Future != 'World']
df_pgr3.head(80)

# Uit de correlatie blijkt dat er een hele sterke correlatie is met de stijgende populatie. De verwachting is dat India het land wordt met de grootste populatie. 
# Daaruit kan je de aanname doen dat India het land wordt met de meeste uitstoot. 
```
